#### Importing all important lib. & dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.stats import chi2_contingency
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve
from sklearn.metrics import precision_recall_curve, average_precision_score, recall_score, precision_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from scipy.stats import f_oneway
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('merged_icu_diabetes_data.csv')
df.head(3)
```

### 1. Modeling the Data: 
• Build a model to predict the likelihood of patient readmission based on the 
combined datasets. Use patient demographics and clinical data such as vital 
signs, lab procedures, and ICU stay information to predict outcomes.  

#### Encode Categorical Variables

```python
#Convert categorical variables like age_group and readmitted into numerical formats.

# Encode the target variable
df['readmitted_binary'] = df['readmitted'].map({'<30': 1, '>30': 1, 'NO': 0})

# One-hot encode categorical features
df_encoded = pd.get_dummies(df, columns=['age_group', 'gender', 'race'], drop_first=True)
df_encoded.head()
```

```python
#Select Features and Target Variable Identify predictors (X) and the target (y).
features = ['Blood_Glucose', 'ICU_Length_of_Stay', 'num_lab_procedures', 
            'num_medications', 'Heart_Rate', 'gender_Male']
X = df_encoded[features]
y = df_encoded['readmitted_binary']
```

### Pridictive Models

```python
#Split the dataset into training and testing sets.

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### Logistic Regrassion Model

```python
# Logistic Regression Start with a simple and interpretable baseline model.
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Make predictions
y_pred_log = log_reg.predict(X_test)
```

#### Random Forest Classifier
- A non-linear model that captures complex patterns.

```python
#Random Forest Build a non-linear, tree-based model.

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

#### Gradient Boosting (XGBoost)
- A high-performance tree-based model.

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)
```

-  Evaluate the model's performance using standard metrics and identify key 
factors influencing predictions.

#### EVALUATION

```python
# Logistic Regression Evaluation
y_pred_log = log_reg.predict(X_test)
print("Logistic Regression Performance:")
print(classification_report(y_test, y_pred_log))
print(f"AUC-ROC: {roc_auc_score(y_test, log_reg.predict_proba(X_test)[:, 1])}")
```

```python
# Random Forest Evaluation
y_pred_rf = rf_model.predict(X_test)
print("\nRandom Forest Performance:")
print(classification_report(y_test, y_pred_rf))
print(f"AUC-ROC: {roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1])}")
```

```python
# XGBoost Evaluation
y_pred_xgb = xgb_model.predict(X_test)
print("\nXGBoost Performance:")
print(classification_report(y_test, y_pred_xgb))
print(f"AUC-ROC: {roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:, 1])}")
```

#### MATRICS EVALUATION

```python
# PREDICTIONS
y_pred = rf_model.predict(X_test)
y_prob = rf_model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# AUC-ROC
auc = roc_auc_score(y_test, y_prob)
print(f"AUC-ROC: {auc:.2f}")
```

### Visulisation of ROC

#### Manual Segmentation Based on Thresholds
- Number of medications: Low (<5), Medium (5–10), High (>10).
- Number of lab tests: Low (<25), Medium (25–50), High (>50).
- ICU stay length: Short (<3 days), Medium (3–7 days), Long (>7 days).

```python
# Define segments
def categorize_medications(num):
    if num < 5:
        return 'Low'
    elif 5 <= num <= 10:
        return 'Medium'
    else:
        return 'High'
```

```python
# Apply to dataset
df['medication_segment'] = df['num_medications'].apply(categorize_medications)
df['lab_tests_segment'] = pd.cut(df['num_lab_procedures'], bins=[0, 25, 50, float('inf')], 
                                  labels=['Low', 'Medium', 'High'])
df['icu_stay_segment'] = pd.cut(df['ICU_Length_of_Stay'], bins=[0, 3, 7, float('inf')], 
                                 labels=['Short', 'Medium', 'Long'])
```

### Statistical Tests
- Statistical tests help determine if the differences in outcomes across segments are significant.
Chi-Square Test for Readmission:

```python
# Create a contingency table
contingency_table = pd.crosstab(df['medication_segment'], df['readmitted_binary'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Test: Chi2 = {chi2}, p-value = {p}")
if p < 0.05:
    print("Significant differences in readmission rates across medication segments.")
else:
    print("No significant differences in readmission rates across medication segments.")
```

```python
# Compare ICU stay lengths across segments
groups = [group['ICU_Length_of_Stay'] for _, group in df.groupby('medication_segment')]
f_stat, p = f_oneway(*groups)
print(f"ANOVA Results: F-Statistic = {f_stat}, p-value = {p}")
if p < 0.05:
    print("Significant differences in ICU stay duration across medication segments.")
else:
    print("No significant differences in ICU stay duration across medication segments.")
```

### Use Predictive Models to Identify Key Risk Factors
Build a predictive model to analyze how segment features influence outcomes like readmission or ICU stays.

Logistic Regression for Readmission:

```python
# Prepare features and target
X = pd.get_dummies(df[['num_medications', 'num_lab_procedures', 'ICU_Length_of_Stay', 'medication_segment']], drop_first=True)
y = df['readmitted_binary']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train logistic regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Evaluate the model
y_pred = log_reg.predict(X_test)
print(classification_report(y_test, y_pred))
```
